### Step 1: Loading all required libraries

```python
# First, we import all the essential libraries needed for this task

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Loading the dataset

```python
# We then load the Iris dataset

df = pd.read_csv("Iris.csv")
df
```

### Step 3: Cleaning and preparing the data for analysis

```python
# The Id column must be dropped to avoid confusion

df.drop(columns = "Id", inplace = True)
df
```

```python
# The iloc() function is then used to store the required rows and columnns from the dataset by creating an array

iris_df = df
x = df.iloc[:,0:4].values
x
```

### Step 4: Determining the optimum number of clusters for K-Means  

```python
# We then find the optimal number of clusters for K-Means and the value of K is determined
# We use 'Within Cluster Sum of Squares' (WCSS) method to help find the optimum K value

from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

### Step 5: Plotting of graph to find optimum k value

```python
# We then plot a graph with WCSS and k values as axes and then use the Elbow method 
# This graph will show the optimum k value to be used
# This is found where the elbow occurs (when WCSS doesn't decrease significantly with increase in number of clusters)

plt.plot(range(1,11),wcss, color='red',marker='X', markerfacecolor='black')
plt.title('Elbow Method')
plt.xlabel('Number Of Clusters')
labels = ["Number Of Clusters" , "WCSS"]
plt.ylabel('WCSS')
plt.legend(labels=labels)
plt.show()
```

### Step 6: Creation of the K-Means classifier

```python
# It is clearly observed that we can choose the number of clusters as '3' since that is where elbow seems to occur
# We then create the KMeans classifier and deploy the KMeans algorithm to the prepared Iris dataset

kmeans = KMeans(n_clusters=3, init="k-means++", max_iter=300, n_init=10, random_state=0)
recognized_clusters = kmeans.fit_predict(x)
recognized_clusters
```

### Step 7: Visualizing the cluster centroids

```python
# We first have a look at the visualization of the plotting the centroids of the clusters

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'indigo', label = 'Centroids')
plt.legend(loc=4)
```

### Step 8: Visualizing the cluster data points

```python
# We then try to visualize the clusters, here done on the first two columns

plt.scatter(x[recognized_clusters  == 0, 0], x[recognized_clusters  == 0, 1], s = 100, c = 'orange', label = 'Iris-setosa')
plt.scatter(x[recognized_clusters  == 1, 0], x[recognized_clusters  == 1, 1], s = 100, c = 'magenta', label = 'Iris-versicolour')
plt.scatter(x[recognized_clusters  == 2, 0], x[recognized_clusters  == 2, 1], s = 100, c = 'lime', label = 'Iris-virginica')
plt.legend(loc=4)
```

### Step 9: Final combination of the 2 graphs

```python
plt.scatter(x[recognized_clusters  == 0, 0], x[recognized_clusters  == 0, 1], s = 100, c = 'orange', label = 'Iris-setosa')
plt.scatter(x[recognized_clusters  == 1, 0], x[recognized_clusters  == 1, 1], s = 100, c = 'magenta', label = 'Iris-versicolour')
plt.scatter(x[recognized_clusters  == 2, 0], x[recognized_clusters  == 2, 1], s = 100, c = 'lime', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'indigo', label = 'Centroids')
plt.legend(loc=4)
```
